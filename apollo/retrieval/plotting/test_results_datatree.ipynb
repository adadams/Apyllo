```python
from xarray import load_dataset, Dataset
```

```python
contributions_dataset: Dataset = load_dataset(
    test_results_directory / "2M2236.Piette.HK+G395H.cloudy.contributions-retooled.nc"
)

samples_dataset: Dataset = load_dataset(
    test_results_directory / "2M2236.Piette.HK+G395H.cloudy.samples.nc"
)

TP_dataset: Dataset = load_dataset(
    test_results_directory / "2M2236.Piette.HK+G395H.cloudy.TP-profiles.nc"
)
```

```python
from apollo.spectrum.read_spectral_data_into_xarray import read_APOLLO_data_into_dataset


MLE_spectrum: Dataset = read_APOLLO_data_into_dataset(
    test_results_directory
    / "2M2236.Piette.HK+G395H.cloudy.resumed-from-checkpoint.Spectrum.binned.dat",
    band_names=["H", "Ks", "G395H_NRS1", "G395H_NRS2"],
)
```

```python
from datatree import DataTree

results_tree_structure: dict[str, Dataset] = {
    "contributions": contributions_dataset,
    "samples": samples_dataset,
    "TP": TP_dataset,
    "MLE": MLE_spectrum,
}

results_dataset: DataTree = DataTree.from_dict(results_tree_structure)
```
